```python
data.PoolQC = data["PoolQC"].fillna("None")
data.Alley = data["Alley"].fillna("None")
data.MiscFeature = data["MiscFeature"].fillna("None")
data.Fence = data["Fence"].fillna("None")
data.MasVnrArea = data["MasVnrArea"].fillna("None")
data.LotFrontage = data["LotFrontage"].fillna("None")
data.MasVnrType = data["MasVnrType"].fillna("None")
data.GarageType = data["GarageType"].fillna("None")
data.GarageCond = data["GarageCond"].fillna("None")
data.GarageQual = data["GarageQual"].fillna("None")
data.GarageFinish = data["GarageFinish"].fillna("None")
data.GarageYrBlt = data["GarageYrBlt"].fillna("None")
data.FireplaceQu = data["FireplaceQu"].fillna("None")
data.BsmtFinType1 = data["BsmtFinType1"].fillna("None")
data.BsmtCond = data["BsmtCond"].fillna("None")
data.BsmtQual = data["BsmtQual"].fillna("None")
data.BsmtFinType2 = data["BsmtFinType2"].fillna("None")
data.BsmtExposure = data["BsmtExposure"].fillna("None")
data.Electrical = data["Electrical"].fillna("None")
```

```python
data.head()
```

```python
data.isnull().any()
```

```python
X = data.drop(['Id','SalePrice'],axis=1)
dm_X = pd.get_dummies(X)
print("X shape: {}".format(X.shape))
print("X with dummies shape: {}".format(dm_X.shape))
y = data['SalePrice']
```

```python
import pickle
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.externals import joblib
from sklearn.svm import SVR

sc = MinMaxScaler()
dm_X = sc.fit_transform(dm_X)
X_train, X_test, y_train ,y_test = train_test_split(dm_X, y,test_size=0.25, random_state= 42,shuffle=True)
```

```python
clf = RandomForestRegressor()
clf.fit(X_train,y_train)
```

```python
print('Accuracy on the training subset: {:.3f}'.format(clf.score(X_train, y_train)))
print('Accuracy on the testing subset: {:.3f}'.format(clf.score(X_test, y_test)))
```

```python
filename = 'housing_model.pkl'
joblib.dump(clf,filename)
```

```python
from sklearn.externals import joblib
mdl = joblib.load('housing_model.pkl')
```

```python
y_pred = mdl.predict(X_test)
```
