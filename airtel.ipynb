```python
import pandas as pd
import seaborn as sns
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib import style
import warnings
warnings.filterwarnings('ignore')

# for elegant plots
style.use('fivethirtyeight')

airtel = pd.read_csv('nasa.csv', dtype={'IMEI':object,'TO_NUMBER':object})

airtel.replace({"DIRECTION": {"OUTGOING\n": "OUTGOING","Senyage": "Senyange"},"CALL TYPE":{"voice":"Voice"}}, inplace=True)
top_towers = airtel.groupby('SITE_NAME').filter(lambda x: len(x) >= 3)
bottom_towers = airtel.groupby('SITE_NAME').filter(lambda x: (len(x) <= 2))

top_towers.groupby('SITE_NAME').size().sort_values().plot.barh(figsize=(15,10))
plt.title("Classified")
plt.ylabel("CTs")
plt.xlabel("Number of calls");
```

```python
gility = gility.dropna()
print(gility.isnull().sum())
import datetime as dt
gility = airtel.loc[~airtel['START_DATE_TIME'].str.contains('\/')]
gility['START_DATE_TIME'] = pd.to_datetime(gility['START_DATE_TIME'], format='%Y-%m-%d %H:%M:%S',errors = 'coerce')
print(gility.dtypes)
gility["day_of_week"] = gility['START_DATE_TIME'].dt.day_name()
gility['DATE'] = [d.date() for d in gility['START_DATE_TIME']]
gility['TIME'] = gility['START_DATE_TIME'].dt.time
gility["MONTH"] = gility["START_DATE_TIME"].dt.month
gility = gility.set_index('DATE')
#gility['MONTH'] = gility.index.month
print(gility.isnull().sum())
gility.head()
```

```python
airtel_incoming = airtel.loc[airtel['DIRECTION'] == 'OUTGOING']
airtel_incoming_may = airtel_incoming.loc[airtel_incoming['START_DATE_TIME'].str.contains('2019-05',regex=True)]
airtel_incoming_june = airtel_incoming.loc[airtel_incoming['START_DATE_TIME'].str.contains('2019-06',regex=True)]
airtel_incoming_july = airtel_incoming.loc[airtel_incoming['START_DATE_TIME'].str.contains('2019-07',regex=True)]
airtel_incoming.sort_values(['DURATION'],ascending=False,inplace=True)

airtel_incoming_may = airtel_incoming_may.groupby('TO_NUMBER').filter(lambda x: len(x) >= 5)
airtel_incoming_june = airtel_incoming_june.groupby('TO_NUMBER').filter(lambda x: len(x) >= 5)
airtel_incoming_july = airtel_incoming_july.groupby('TO_NUMBER').filter(lambda x: len(x) >= 5)
```
