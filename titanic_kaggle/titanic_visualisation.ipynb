```python
# My custom class wherein are libraries like pandas ,mpl,np etc.
from preprocessor import *
prep = Preprocessor()

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
td = pd.concat([train,test],ignore_index=True,sort=False)
td['Pclass'] = td['Pclass'].astype(object)
#prep.overview(train)
```

```python
tn = td.isnull()
plt.figure(figsize=(15,8))
sns.heatmap(tn,cbar=False).set_title("Missing values heatmap");
```

```python
with_cabin = train.loc[~train['Cabin'].isnull()]
print(f'People with cabin: {with_cabin.shape[0]}')

plt.figure(figsize=(10,6))
sns.countplot(x="Survived",data=with_cabin)
plt.ylabel('Number of people')
#show the plot
plt.show()
```

```python
no_cabin = train.loc[train['Cabin'].isnull()]
print(f'People with no cabin: {no_cabin.shape[0]}')

plt.figure(figsize=(12,8))
sns.countplot(x="Survived",data=no_cabin)
plt.ylabel('Number of people');
```

```python
plt.figure(figsize=(10,6))
sns.pointplot(x="Age",data=td)
plt.xticks()
#show the plot
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Pclass",data=td)
plt.xticks()
#show the plot
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Pclass",hue='Survived',data=train)
plt.xticks()
#show the plot
plt.show()
```

```python
FacetGrid = sns.FacetGrid(train, row='Embarked', size=4.5, aspect=1.6)
FacetGrid.map(sns.pointplot, 'Pclass', 'Survived', 'Sex', order=None, hue_order=None )
FacetGrid.add_legend();
```

```python
grid = sns.FacetGrid(train, col='Survived', row='Pclass', size=3.2, aspect=1.6)
grid.map(plt.hist, 'Age', alpha=.5, bins=20)
grid.add_legend();
```
